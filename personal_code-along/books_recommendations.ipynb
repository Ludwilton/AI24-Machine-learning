```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
users = pd.read_csv("../data/books/Users.csv")
books = pd.read_csv("../data/books/Books.csv")
ratings = pd.read_csv("../data/books/Ratings.csv")
```

```python

print(users.info)
print(ratings.info)
print(books.info)
```

```python
books.duplicated()
books[books["Book-Title"].duplicated()].sort_values("Book-Title")

new_books = books.drop_duplicates("Book-Title")
new_books.info()
```

```python
ratings_with_name = ratings.merge(new_books, on="ISBN")
ratings_with_name.drop(["ISBN", "Image-URL-S","Image-URL-M","Image-URL-L", "Publisher"], axis = 1, inplace=True)

ratings_with_name
```

```python
users_ratings_matrix = ratings_with_name.merge(users,on="User-ID")

users_ratings_matrix
```

```python
print(ratings_with_name.isna().sum())
ratings_with_name.dropna(inplace=True)
print(ratings_with_name.isna().sum())
ratings_with_name.info()
```

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"]>100
expert_users = x[x].index

filtered_users_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(expert_users)]

y = filtered_users_ratings.groupby("Book-Title").count()["Book-Rating"]>= 50
famous_books= y[y].index
user_ratings = filtered_users_ratings[filtered_users_ratings["Book-Title"].isin(famous_books)]
```

```python
design_matrix = user_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
design_matrix.fillna(0,inplace=True)
design_matrix
```

```python
scaler = StandardScaler(with_mean=True, with_std=True)
design_matrix_centered = scaler.fit_transform(design_matrix)
similarity_score = cosine_similarity(design_matrix_centered)
```

```python
def recommend(book_name):
    

    index = np.where(design_matrix.index==book_name)[0][0]
    

    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    

    data = []
    
    for index,similarity in similar_books:
        item = []

        temp_df = new_books[new_books['Book-Title'] == design_matrix.index[index]]
        

        item.extend(temp_df['Book-Title'].values)
        item.extend(temp_df['Book-Author'].values)
        
        data.append(item)
    return data
```

```python
recommend('2010: Odyssey Two')
```
